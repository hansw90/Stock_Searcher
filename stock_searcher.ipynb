# 종목 검색기

내가 원하는 조건의 주식을 찾아내고 csv 파일로 저장해보자,

```python
# module import 

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas_datareader import data as web
from tqdm.notebook import tqdm

%matplotlib inline
```

```python
# stock list load

# 상장법인 리스트를 받아옵니다.
code_df = pd.read_html('http://kind.krx.co.kr/corpgeneral/corpList.do?method=download&searchType=13', header=0)[0]

code_df.종목코드 = code_df.종목코드.map('{:06d}'.format)

code_df = code_df[['회사명','종목코드']]

code_df = code_df.rename(columns={'회사명':'name','종목코드':'code'})
code_df.head()
```

```python
len(code_df)
```

```python
stock_codes = list(code_df['code'])
```

```python
def get_url(code) :
    url = 'http://finance.naver.com/item/sise_day.nhn?code={code}'.format(code=code)
    url = ''.join(url.split(' '))
    
    return url
```

```python
# request url 
stock_urls = list(map(get_url, stock_codes))
stock_urls[:5]
```

```python
# 이제 위 stock_urls 를 통해 주식정보를 가져온다,

def get_stock_info(url) :
    
    df = pd.DataFrame()
    for page in range(1, 21):

        pg_url = '{url}&page={page}'.format(url=url, page=page)
        df = df.append(pd.read_html(pg_url, header=0 , index_col=False)[0],  ignore_index=True)
        
    return df
```

```python
# for문을 돌며 검색을 한다.

new_df = pd.DataFrame(columns={'name','code'})

for idx,url in tqdm(enumerate(stock_urls)) :
    df = get_stock_info(url)
    df = df.dropna()
    
    # 한글로 된 컬럼명을 영어로 바꿔줌
    df = df.rename(columns= {'날짜': 'date', '종가': 'close', '전일비': 'diff', 
                             '시가': 'open', '고가': 'high', '저가': 'low', '거래량': 'volume'})

    # 데이터의 타입을 int형으로 바꿔줌
    df[['close', 'diff', 'open', 'high', 'low', 'volume']]
        = df[['close', 'diff', 'open', 'high', 'low', 'volume']].astype(int)

    # 컬럼명 'date'의 타입을 date로 바꿔줌
    df['date'] = pd.to_datetime(df['date'])

    # 일자(date)를 기준으로 오름차순 정렬
    df = df.sort_values(by=['date'], ascending=True)

    df['30 Day MA'] = df['close'].rolling(window=20).mean()
    df['30 Day STD'] = df['close'].rolling(window=20).std()
    df['Upper Band'] = df['30 Day MA'] + (df['30 Day STD'] * 2)
    df['Lower Band'] = df['30 Day MA'] - (df['30 Day STD'] * 2)
    
    
    df = df.sort_values(by=['date'], ascending=True, ignore_index=True)

    
    # 너무 지저분한데??...
    if (df[-1:]['close'] <= df[-1:]['Lower Band']).to_string(index = False) != ' False' :
        
        new_df = new_df.append(code_df[idx:idx+1])
        
        plt.title(code_df["code"][idx])
        plt.plot(df[['close', 'Upper Band', 'Lower Band']])
        plt.figure(figsize=(50, 10))
        plt.show()
```

```python
# 계속 추가 하지 않고 덮어써 사용하겠습니다.

new_df.to_csv("stocks.csv", mode='w')
```
